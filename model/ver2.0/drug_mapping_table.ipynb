```python
import pandas as pd
import numpy as np
```

```python
df_ingredient = pd.read_excel("/content/drugtop_ingredient.xlsx")
df_ingredient
```

```python
df_train = pd.read_csv("https://raw.githubusercontent.com/SNMHZ/Drug_Recommendation/master/dataset/1/drugsComTrain_raw.csv").dropna().reset_index()
df_train
```

```python
df_train.drop(df_train[df_train['condition'].str.contains('</span>')].index, axis=0, inplace=True)
```

```python
df_train = df_train[df_train['drugName'].isin(df_ingredient['drugName'].unique())].reset_index()
df_train = df_train[['drugName', 'condition', 'rating', 'usefulCount']]
df_train
```

```python
map_drug_ingredient = df_ingredient[['drugName', '성분']].set_index('drugName')
map_drug_ingredient
```

```python
df_train['성분'] = df_train['drugName'].apply(lambda x: map_drug_ingredient.loc[x, '성분'])
df_train
```

```python
condition_comp = {'Anxiety and Stress':'Anxiety',
'Weight Loss':'Obesity'}
```

```python
def cond_compress(x):
  if x in condition_comp:
    return condition_comp[x]
  return x
```

```python
df_train['condition'] = df_train['condition'].apply(cond_compress)
```

```python
pd.read_excel("top20condition.xlsx")['condition']
```

```python
df_train = df_train[df_train['condition'].isin( pd.read_excel("top20condition.xlsx")['condition'] )]
```

```python
df_train
```

```python
df_train[df_train['condition']=='Anxiety']
```

```python
table = {}
for condition in df_train['condition'].unique():
  df_con = df_train[df_train['condition']==condition]
  con_list = []
  for ingred in df_con['성분'].unique():
    con_list.append( (ingred, df_con[df_con['성분']==ingred]['rating'].sum()/df_con[df_con['성분']==ingred].count()['rating']) )
  table[condition]=sorted(con_list, key=lambda x: x[1], reverse=True)
```

```python
mapping_table = pd.DataFrame(table.items()).rename(columns={0:'condition', 1:'drugs'})
```

```python
mapping_table.set_index('condition').to_pickle("mapping_table.pickle")
```
