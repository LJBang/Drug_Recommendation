# TextBlob (자연어 처리 라이브러리)

```python
# The basic imports
import numpy as np
import pandas as pd
import os
from tqdm import tqdm # 반복문의 진행 상황을 알려줌
```

```python
from textblob import TextBlob # 자연어 처리 라이브러리
```

```python
example = "This is a wonderful product. I got amazing results by using it"
blob = TextBlob(example) # create the TextBlob object for this example
```

```python
blob.sentiment
```

sentimet 메서드를 쓰면 polarity와 subjectivity 점수가 나오는데

먼저 polarity는 -1 ~ 1부터의 범위를 갖는다.

1에 가까울수록 긍정적인 리뷰이다.

subjectivity는 0 ~ 1의 범위를 갖는다.

1에 가까울수록 주관적인 리뷰이다. (0 -> 사실, 1 -> 의견)

```python
example = "This is a terrible product. I would not recommend using it"
blob = TextBlob(example)
blob.sentiment
```

```python
example = "Jupiter is the largest planet in the Solar System"
blob = TextBlob(example)
blob.sentiment
```

이번 분석에서는 polarity가 훨씬 더 중요하다.


# 데이터 분석

```python
train = pd.read_csv('https://github.com/SNMHZ/Drug_Recommendation/raw/master/dataset/drugsComTrain_raw.csv')

# Inspect first 5 rows
train.head(5)
```

```python
train.shape
```

```python
# 리뷰만 모으기

reviews = train["review"]
print(reviews[:10]) # First 10
```

```python
# 감정 추출을 해보자

sentiments = []
for review in tqdm(reviews):
    blob = TextBlob(review)
    sentiments += [blob.sentiment.polarity]
```

```python
# train 셋에 polarity 결과 칼럼을 추가한다.

train["sentiment"] = sentiments
```

```python
train.sample(10)
```

Sentiment analysis는 짧고 명확한 리뷰에서 가장 명확한 결과를 내는데

리뷰가 길면 rating이 높아도 낮은 감정 결과를 낼 수 있다.

또 중립적인 문장이 많이 들어가면 높은 감정 결과가 희석되어서 낮게 나올 수 있다.

```python
# Correlation with ratings
# ratings 와의 상관 관계 알아보기

np.corrcoef(train["rating"], train["sentiment"])
```

생각보다 낮지만 쓸만하다?

```python
# 좀 더 자세하게 그래프로 그려보자

import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x=np.array(train["rating"]),y=np.array(train["sentiment"]))
plt.xlabel("Rating")
plt.ylabel("Sentiment")
plt.title("Sentiment vs Ratings")
plt.show()
```
